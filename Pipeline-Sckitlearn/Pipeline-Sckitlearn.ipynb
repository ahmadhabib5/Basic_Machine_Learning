# Membuat model dengan pipeline

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('breast-cancer-datasets.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 32','id'], inplace=True)
```

```python
df.head()
```

```python
df_baru = df.copy(deep=True)
```

```python
df_baru['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
df_baru.head()
```

```python
correlation = df_baru.corr()['diagnosis']
correlation
```

```python
strong_corr = correlation[correlation >= 0.5]
strong_corr
```

```python
strong_corr.keys().to_list()
```

```python
y = df_baru['diagnosis']
X = df_baru[strong_corr.keys().to_list()[1:]]
```

```python
X.shape , y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
X_train.dtypes
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
numerical_pipeline = Pipeline([
    ('inputer', SimpleImputer(strategy='mean')),
    ('sclare', MinMaxScaler())
])
```

```python
from sklearn.svm import SVC
```

```python
pipeline = Pipeline([
    ('prepo', numerical_pipeline),
    ('algo', SVC())
])
```

## parameter Tuning

```python
parameter = {
    'algo__C': [1,10,100,1000],
    'algo__degree': range(1,5)
}
```

```python
model = GridSearchCV(pipeline, param_grid=parameter, cv=3, n_jobs=-1, verbose=2)
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.score(X_test, y_test)
```

```python
pd.DataFrame(model.cv_results_).sort_values('rank_test_score')
```
